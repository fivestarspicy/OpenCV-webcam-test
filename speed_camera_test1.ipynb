```python
import cv2

# Set up the camera
cap = cv2.VideoCapture(0)

# Check if the camera is opened successfully
if not cap.isOpened():
    print("Error opening camera")
    exit()

while(True):
    # Capture frame-by-frame
    ret, frame = cap.read()

    # Convert to grayscale
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Apply threshold to create a binary image
    ret, thresh = cv2.threshold(gray,127,255,cv2.THRESH_BINARY)

    # Find contours
    contours, hierarchy = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Draw all contours on the frame
    cv2.drawContours(frame, contours, -1, (0, 255, 0), 2)

    # Display the resulting frame with contours drawn
    cv2.imshow('frame', frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the capture
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

# Set up the camera
cap = cv2.VideoCapture(0)

# Check if the camera is opened successfully
if not cap.isOpened():
    print("Error opening camera")
    exit()

# Set the first frame as the reference frame
_, ref_frame = cap.read()

# Filter for blue pixels in the reference frame
lower_blue = np.array([110, 50, 50])
upper_blue = np.array([130, 255, 255])
ref_mask = cv2.inRange(ref_frame, lower_blue, upper_blue)

# Find contours of the filtered reference frame
contours, hierarchy = cv2.findContours(ref_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Find the largest contour
if len(contours) > 0:
    c = max(contours, key = cv2.contourArea)

    # Draw a bounding rectangle around the largest contour
    x,y,w,h = cv2.boundingRect(c)
    cv2.rectangle(ref_frame,(x,y),(x+w,y+h),(255,0,0),2)

    # Get the center of the bounding rectangle
    obj_x, obj_y = x + w/2, y + h/2
    
    while(True):
        # Capture frame-by-frame
        ret, frame = cap.read()

        # Filter for blue pixels in the current frame
        mask = cv2.inRange(frame, lower_blue, upper_blue)

        # Find contours of the filtered image
        contours, hierarchy = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

        # Find the largest contour
        if len(contours) > 0:
            c = max(contours, key = cv2.contourArea)

            # Draw a bounding rectangle around the largest contour
            x,y,w,h = cv2.boundingRect(c)
            cv2.rectangle(frame,(x,y),(x+w,y+h),(255,0,0),2)

            # Get the center of the bounding rectangle
            new_obj_x, new_obj_y = x + w/2, y + h/2

            # Calculate displacement and speed
            displacement = np.sqrt((new_obj_x - obj_x)**2 + (new_obj_y - obj_y)**2)
            speed = displacement / (1/cap.get(cv2.CAP_PROP_FPS)) # displacement per frame

            # Update the object's position
            obj_x, obj_y = new_obj_x, new_obj_y
        
        # Display the resulting frame
        cv2.imshow('frame', frame)
        cv2.putText(frame, "Speed: {:.2f} pixels/frame".format(speed), (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
# Release the capture
cap.release()
cv2.destroyAllWindows()
```
